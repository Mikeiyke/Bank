```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bank-additional-full.csv', sep=';')
```

```python
# Calculate subscription rate by marital status
married_df = df[df['marital'] == 'married']
single_df = df[df['marital'] == 'single']
```

```python
married_subscription_rate = (married_df['y'] == 'yes').mean()
single_subscription_rate = (single_df['y'] == 'yes').mean()
```

```python
print(f"Married Subscription Rate: {married_subscription_rate:.2f}")
print(f"Single Subscription Rate: {single_subscription_rate:.2f}")
```

```python
# Visualize the results
plt.figure(figsize=(8, 6))
subscription_rates = [married_subscription_rate, single_subscription_rate]
labels = ['Married', 'Single']
plt.bar(labels, subscription_rates)
plt.title('Subscription Rate by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Subscription Rate')
plt.show()
```

## DO CLIENTS WHO WERE PREVIOUSLY CONTACTED (PREVIOUS>0) TEND TO SUBSCRIBE MORE OFTEN?

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('bank-additional-full.csv',sep=';')
```

```python
#calculate subsription rate for clients with previous contacts
previous_contacts=df[df['previous']>0]
no_previous_contacts=df[df['previous']==0]
```

```python
previous_subscription_rate=(previous_contacts['y']=='yes').mean()
no_previous_subscription_rate=(no_previous_contacts['y']=='yes').mean()
```

```python
print(f"subscription rate for clients with previous contacts:{previous_subscription_rate:.2f}")
print(f"subscription rate for clients without previous contacts:{no_previous_subscription_rate:.2f}")
```

```python
#Visualize the results 
plt.figure(figsize=(8,6))
subscription_rates=[previous_subscription_rate, no_previous_subscription_rate]
labels=['previous contacts','no previous contacts']
plt.bar(labels,subscription_rates)
plt.title('subscription rate by previous contacts')
plt.ylabel('subscription rate')
plt.show()
```

## A BARCHART 

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'job': ['admin', 'blue-collar', 'entrepreneur', 'admin', 'student'],
    'marital': ['married', 'single', 'married', 'divorced', 'single'],
    'education': ['high.school', 'basic.4y', 'university.degree', 'high.school', 'basic.9y']
}

df = pd.DataFrame(data)

# Count occurrences of each job category
job_counts = df['job'].value_counts()

# Create bar chart
plt.figure(figsize=(8, 6))
job_counts.plot(kind='bar')
plt.title('Job Categories')
plt.xlabel('Job')
plt.ylabel('Count')
plt.show()
```

## Correlation 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('bank-additional-full.csv', sep=';')

# Encode target variable 'y'
df['y'] = (df['y'] == 'yes').astype(int)

# Select numerical features
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns

# Calculate correlation matrix
corr_matrix = df[numerical_features].corr()

# Visualize correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
